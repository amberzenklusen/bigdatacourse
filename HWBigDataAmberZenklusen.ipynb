```python
#Data is now relatively tidy (following Wickham 2014: 5): Columns are named, each observation forms a row, each type of observational unit forms a table, each column is one variable only. Several variables, however are still nested in a dict-format, and need to be unpacked.  
#Source: Wickham, H., (2014). Tidy Data, Journal of Statistical Software. Foundation for Open Access Statistics, vol. 59(i10).
```

```python
#As 'Answering Body' and 'questionText' are still in a dict format, and need to be unpacked. 
def unnesting(df, explode):
    idx = df.index.repeat(df[explode[0]].str.len())
    df1 = pd.concat([
        pd.DataFrame({x: np.concatenate(df[x].values)}) for x in explode], axis=1)
    df1.index = idx
    return df1.join(df.drop(explode, 1), how='left')

#Source of code: 
#stackoverflow.com/questions/53218931/how-to-unnest-explode-a-column-in-a-pandas-dataframe, user: @BENY (accessed on April 1, 2021).
```

```python
df = unnesting(df,['AnsweringBody','tablingMemberPrinted'])
df
```

```python
res = pd.DataFrame([{'_value': key, 'Body' : value } for d in df['AnsweringBody'].tolist() for key, value in d.items()])
print(res)
df2 = pd.DataFrame([{'_value': key, 'MP': value } for d in df['tablingMemberPrinted'].tolist() 
    for key, value in d.items()])
df = pd.concat([df,res,df2], axis = 1)
df
```

```python
#Very basic text analysis of Questions to inspect text. 
#Extracted from dataframe for a separate analysis. Depending on the question, however, the questions could also be analyzed in conjuction with other variables (e.g constituency label or MP)
from collections import Counter 
```

```python
corpus = df['questionText'].tolist()
corpus
```

```python
import nltk
nltk.download('punkt') # for sent_tokenize
nltk.download('stopwords') 
nltk.download('wordnet') # for WordNetLemmatizer
```

```python
c = Counter(corpus)
c.most_common(10)
#some questions were submited several times (up to 5 times) / some requests are general request and exact question is unknown (max 50 times)
```

```python
#get only unique questions for text analysis to understand what topic were important 
output = []
for x in corpus:
    if x not in output:
        output.append(x)
print(output)
```

```python
import re 
from nltk.tokenize import TreebankWordTokenizer
from nltk.tokenize import WhitespaceTokenizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
output = ''.join(str(e) for e in corpus) #join list of strings into one string for shared analysis. 
```

```python
from nltk.tokenize import wordpunct_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
fd = FreqDist( #find out most frequent words in questions (unique questions)
    word
    for word in wordpunct_tokenize(output)
    if word not in set(stopwords.words('english'))
)
fd.pprint(200)
#some of the most important topics are 'trade', 'health', 'covid', while 'crisis', 'tariffs', 'businesses' are surprisingly less frequent. 
```
